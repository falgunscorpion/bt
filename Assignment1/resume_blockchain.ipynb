```python
import fitz
import os
import re
```

```python
class Resume:
    def __init__(self,filename,text):
       self.filename = filename
       self.text = text

students = []
        
```

```python
def extract_text(filename):
    text = ""
    doc = fitz.open(filename)
    for page in doc:
        text = text + page.get_text()
    return text
```

```python
folder_path = "./resume"
filenames = os.listdir(folder_path)
print(filenames)
```

```python
for filename in filenames:
    text = extract_text(f"{folder_path}/{filename}")
    student = Resume(filename,text)
    students.append(student)
```

```python
for student in students:
    print(student.filename)
    print(student.text)
    email = re.search(r'\S+@\S+', student.text).group()
    print(email)
```

```python
import hashlib
```

```python
class Block:
  def __init__(self, data, prev_hash):
    self.data = data
    self.prev_hash = prev_hash
    self.hash = self.calc_hash()

  def calc_hash(self):
    sha = hashlib.sha256()
    sha.update(self.data.encode('utf-8'))
    return sha.hexdigest()
```

```python
class Blockchain:

  def __init__(self):
    self.chain = [self.create_genesis_block()]

  
  def create_genesis_block(self):
    return Block("This is the genesis block", "0")

  
  def add_block(self, data):
    prev_block = self.chain[-1]
    new_block = Block(data, prev_block.hash)
    self.chain.append(new_block)

```

```python
blockchain = Blockchain()
```

```python
for student in students:
    blockchain.add_block(student.text)
```

```python
print("Blockchain:")

for block in blockchain.chain:
    print('Data :',block.data)
    print('Privious Hash :', block.prev_hash)
    print('Hash :', block.hash)
    print("**********************************************************************************************************************************************************************")
```
